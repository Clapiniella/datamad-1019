```python
import pandas as pd

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
df = pd.read_csv('data/vehicles/vehicles.csv')
display(df.head())
print(df.columns)
print(df.shape)
```

```python
df.dtypes
```

```python
# Casting de toda la columna
df["Year"] = df["Year"].astype(int)
```

```python
df = df.rename(columns={"Make": "Manufacturer"})
```

```python
filtered = df[(df['Manufacturer']=='Ford') & (df['Cylinders']>=6) & (df['Combined MPG'] < 18)]

print(filtered.shape)
filtered.head()
```

```python
#df["Year"].value_counts().sort_values()
df["Year"].value_counts().sort_index()
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.cut.html?highlight=cut#pandas.cut
cutoffs = [7,14,21,23,30,40]
mpg_labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']
bins = pd.cut(df['Combined MPG'],cutoffs, labels=mpg_labels)
df["consumo"] = bins
```

```python
df["consumo"].value_counts()
```

```python
df.loc[df['Transmission'].str.startswith('A'), 'TransType'] = 'Automatic'
df.loc[df['Transmission'].str.startswith('M'), 'TransType'] = 'Manual'
```

```python
df["TransType"].value_counts()
```

```python
top_marcas = df.groupby('Manufacturer',as_index=False).agg({'Combined MPG':'mean',"Year":'min'})
#top_marcas.reset_index()
top_marcas[(top_marcas["Manufacturer"]=="Ford")| (top_marcas["Manufacturer"]=="Audi")]
#avg_mpg.columns = ['Make', 'Avg_MPG']
```

```python
avg_mpg = df.groupby('Manufacturer', as_index=False).agg({'Combined MPG':'mean'})
avg_mpg.columns = ['Manufacturer', 'Avg_MPG']
display(avg_mpg)
df2 = pd.merge(df, avg_mpg, on='Manufacturer')
df2.head(10)
```

```python
melted = pd.melt(df, id_vars=['Year','Manufacturer','Model'], 
                 value_vars=['City MPG'])
melted.head(10)
print(melted)
print(melted.shape)
```
