```python
def square(L=1, offset=[0,0], autocenter=True):
    if autocenter:
        offset = [-L/2,-L/2]
    c = np.asarray([
        [0,0],
        [0,L],
        [L,L],
        [L,0],
        [0,0]
    ])
    return c[:,0]+offset[0], c[:,1]+ offset[1]

plt.figure(figsize=(5,5))
for i in range(10):
    plt.plot(*square(i), '*-')
```

```python
plt.figure(figsize=(5,5))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(*square(i), '*-')
    plt.plot(*square(i*2), '*-')
```

```python
# https://es.wikipedia.org/wiki/Matriz_de_rotaci%C3%B3n
import math

plt.figure(figsize=(5,5))
for i in range(10):
    degrees = 5
    a = i*degrees * (2*math.pi/360)
    rt = [[math.cos(a), -math.sin(a)],
         [math.sin(a), math.cos(a)]]
    c = np.dot(np.asarray(square(i)).T,rt)
    plt.plot(*c.T, '*-', color='blue', alpha=1.0/(i+1))
```

```python
ventas = [100,500,200,150,800]
meses = [1,2,3,4,5]

plt.bar(meses,ventas)
plt.ylabel('Tomates')
plt.xlabel('Mes')
plt.xticks(meses, labels=['Enero','Febrero','Marzo','Abril','Mayo'], rotation=45)
plt.legend(["Venta tomates mes"])
plt.title("Prevision de venta de tomates Marc")

plt.annotate('Vendimos muchos\ntomates maduros', xy=(2, ventas[1]), xytext=(3, ventas[1]*1.5),
            arrowprops=dict(facecolor='#00FF00', shrink=0.05))
```

```python
plt.figure()
with plt.xkcd():
    ventas_tomates = np.array([100,500,200,150,800])
    ventas_naranjas = np.flip(np.array([100,500,200,150,800]))

    meses = np.array([1,2,3,4,5])

    plt.bar(meses-0.1,ventas_tomates, 0.2)
    plt.bar(meses+0.1,ventas_naranjas,0.2)

    plt.ylabel('Tomates')
    plt.xlabel('Mes')
    plt.xticks(meses, labels=['Enero','Febrero','Marzo','Abril','Mayo'], rotation=45)
    plt.legend(["Tomates","Naranjas"])
    plt.title("Prevision de venta de tomates Marc")

    plt.annotate('Vendimos muchos\ntomates maduros', xy=(2, ventas_tomates[1]), xytext=(3, ventas_tomates[1]*1.1),
                arrowprops=dict(facecolor='#00FF00', shrink=0.05))
```

```python
import pandas as pd
ventas_tomates = np.array([100,500,200,150,800])
ventas_naranjas = np.flip(np.array([100,500,200,150,800]))
meses = np.array(['Enero','Febrero','Marzo','Abril','Mayo'])

df = pd.DataFrame(ventas_tomates, index=meses, columns=["Tomates"])
df["Naranjas"] = ventas_naranjas
ax = df.plot.bar()

ax.annotate('Vendimos muchos\ntomates maduros', xy=(1, ventas_tomates[1]), xytext=(2, ventas_tomates[1]*1.1),
            arrowprops=dict(facecolor='#00FF00', shrink=0.05))
```
