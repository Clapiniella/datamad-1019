```python
import requests


res = requests.get("https://es.wikipedia.org/wiki/Juegos_Ol%C3%ADmpicos_de_Barcelona_1992")
html = res.text
html[:20]
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
tabla_medallero = soup.find_all('table')[-4]
```

```python
filas = tabla_medallero.find_all('tr')[1:]
medallero = []
for fila in filas:
    celdas = fila.find_all('td')
    medallas = {
        "pais": celdas[1].find('a').text.strip(),
        "oro": int(celdas[2].text),
        "plata": int(celdas[3].text),
        "bronce": int(celdas[4].text)
    }
    medallero.append(medallas)
    
print(medallero)
```

```python
sum(map(lambda pais: pais["oro"],medallero))
```

```python
span = soup.select(".mw-headline")[0]
print(span)
print(span.text)
print(span.get('class'))
print(span.attrs)
```
